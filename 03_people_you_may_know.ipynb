# Finding "People You May Know"
Now that our data is cleaned and structured, your manager assigns you a new task: Build a 'People You May Know' feature!

In social networks, this feature helps users connect with others by suggesting friends based on mutual connections. Your job is to analyze mutual friends and recommend potential connections.

# Task 1: Understand the Logic
## How 'People You May Know' Works:
- If User A and User B are not friends but have mutual friends, we suggest User B to User A and vice versa.
- More mutual friends = higher priority recommendation.
### Example:

- Amit (ID: 1) is friends with Priya (ID: 2) and Rahul (ID: 3).
- Priya (ID: 2) is friends with Sara (ID: 4).
- Amit is not directly friends with Sara, but they share Priya as a mutual friend.
- Suggest Sara to Amit as "People You May Know".
But there are cases where we will have more than one "People You May Know". In those cases, greater the number of mutual friends, higher the probability that the user might know the person we are recommending.

# Task 2: Implement the Algorithm
We'll create a function that:

- Finds all friends of a given user.
- Identifies mutual friends between non-friends.
- Ranks recommendations by the number of mutual friends.

```python
import json  # Import the JSON module to work with JSON files

# Load the data from a JSON file
def load_data(file):
    # Open the file in read mode
    with open(file,"r") as f:
        # json.load() converts JSON data into a Python dictionary
        return json.load(f)


# Function to recommend "People You May Know"
# user_id -> the user for whom we want friend suggestions
# data -> the data loaded from the JSON file
def people_you_may_know(user_id, data):

    # Dictionary to store each user's friends
    # Format: {user_id: set_of_friends}
    user_friends = {}

    # Loop through each user in the dataset
    for user in data["users"]:
        # Convert friend list into a set to remove duplicates
        user_friends[user["id"]] = set(user["friends"])  # data cleaning

    # If the given user does not exist in the dataset
    if user_id not in user_friends:
        return []

    # Get the direct friends of the user
    direct_friends = user_friends[user_id]

    # Dictionary to store suggested users and
    # the number of mutual friends they share
    # Format: {suggested_user_id: mutual_friend_count}
    suggestions = {}


    for friend in direct_friends:
        # For every friend, check their friends
        # (these are "friends of friends")
        for mutual in user_friends[friend]:

            # Suggest only if:
            # 1. The mutual friend is not the same as the user
            # 2. The mutual friend is not already a direct friend
            if mutual != user_id and mutual not in direct_friends:

                # Count the number of mutual friends
                # suggestions.get(mutual, 0) returns the current count
                # If the user does not exist yet in the dictionary,
                # it starts counting from 0
                suggestions[mutual] = suggestions.get(mutual, 0) + 1


    # Sort the suggested users based on number of mutual friends
    # x[1] refers to the mutual friend count
    # reverse=True sorts from highest to lowest
    sorted_suggestions = sorted(
        suggestions.items(),
        key=lambda x: x[1],
        reverse=True
    )

    # Extract only the user IDs from the sorted list
    # Ignore the mutual friend counts
    return [user_id for user_id, _ in sorted_suggestions]



# -----------------------------
# Main Program Execution
# -----------------------------

# Load the JSON data file
data  = load_data("massive_data.json")

# Define the user for whom we want recommendations
user_id = 8

# Generate friend recommendations
recommendations = people_you_may_know(user_id, data)

# Print the recommended user IDs
print(f"People you may know for user {user_id}:{recommendations}")
```
