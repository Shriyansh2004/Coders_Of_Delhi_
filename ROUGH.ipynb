```python
import json
from typing import List

from traitlets import List

# DATA LOADING:
def load_data(filename):
    with open(filename, 'r') as file:
        return json.load(file)

# PEOPLE YOU MAY KNOW:
def People_you_may_know(user_id, data):
    user_friends = {}

    # 1st we have to find who are the friends of the user
    for user in data["users"]:
        user_friends[user["id"]] = set(user['friends'])
    if user_id not in user_friends:
        return []
    # after finding: store it
    direct_friends = user_friends[user_id]

    suggestions = {}
    # Loop to find friends in direct friends
    for friend in direct_friends:
        # Loop to find mutual
        for mutual in user_friends[friend]:
            # If mutual not found
            if mutual != user_id and mutual not in direct_friends:
                # Then give suggestions
                suggestions[mutual] = suggestions.get(mutual, 0)+1

    # Sorting the Suggestions
    sorted_suggestions = sorted(
        suggestions.items(),
        key=lambda x: x[1],
        reverse=True
    )
    # After end return the userid with his/her suggestions
    return [user_id for user_id, _ in sorted_suggestions]


# PAGE YOU MIGHT LIKE:
def Page_you_might_like(user_id, data):
    user_pages = {}

    # 1st we have to find which pages are liked by the user
    for user in data["users"]:
        user_pages[user["id"]] = set(user['liked_pages'])
    if user_id not in user_pages:
        return []
    # after finding: store it
    user_liked_pages = user_pages[user_id]

    page_suggestions = {}
    # Loop  for finding pages between user page and other user pages
    for other_user, pages in user_pages.items():
            # If not found
            if other_user != user_id:
                # then find intersection;  for suggestions
                shared_pages = user_liked_pages.intersection(pages)
                for page in pages:
                    if page not in shared_pages:
                        page_suggestions[page] = page_suggestions.get(page, 0)+len(shared_pages)
    # Sorting the Suggestions
    sorted_page_suggestions = sorted(
        page_suggestions.items(),
        key=lambda x: x[1],
        reverse=True
    )
    # After end return the userid with his/her suggestions
    return [page_id for page_id, _ in sorted_page_suggestions]



data  = load_data("massive_data.json")
user_id = 4
people_recommendations = set(People_you_may_know(user_id, data))
print(f"People you may know for user {user_id}:{people_recommendations}")
page_recommendations = set(Page_you_might_like(user_id, data))
print(f"Page you might like{user_id}:{page_recommendations}")
```
